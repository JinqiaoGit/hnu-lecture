```python
import requests
import lxml
import json
from bs4 import BeautifulSoup

html =  """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1"><!--Elsie--></a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>
<p class="story">...</p>
</body>
"""

soup = BeautifulSoup(html,'html.parser')

#1、向下遍历
print(soup.p.contents)
print(list(soup.p.children))
print(list(soup.p.descendants))
print('*'*20)

"""
.contents 返回的是一个 (子节点)的列表，将<tag>所有儿子节点存入列表 。
.children返回的是一个 遍历 (子节点)的迭代器，与.contents类似，用于循环遍历儿子节点。
.descendants 返回的是一个 (子孙节点)的迭代器，包含所有子孙节点，用于循环遍历。
"""

#2、向上遍历
print(soup.p.parent.name,'\n')
for i in soup.p.parents:
    print(i.name)
print('*'*20)

#3、平行遍历
print('a_next:',soup.a.next_sibling) # tag标签下一兄弟节点
for i in soup.a.next_siblings: # tag标签后续全部兄弟节点
    print('a_nexts:',i)
print('a_previous:',soup.a.previous_sibling) #tag标签上一兄弟节点
for i in soup.a.previous_siblings: #tag标签前序全部兄弟节点
    print('a_previouss:',i)
```

## 12. 多线程和多进程

```python
import time
from threading import Thread
from multiprocessing import Process
```

```python
def func():
    for item in range(30):
        print("func1", item)
```

```python
# case 1：多线程
t = Thread(target=func)
t.start()
for i in range(1000):
    print("main", i)
```

```python
def func2(key='abc'):
    for item in range(30):
        print(f"Process1_{key}_", item)
```

```python
# case 2：进程
p = Process(target=func)
p.start()
for i in range(20):
    print("main", i)
```

```python
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor

with ThreadPoolExecutor(50) as t:
    for i in range(100):
        t.submit(func2(), key=f"线程{i}")
    # 等待线程池中的任务全部执行完成，才继续执行
    print('end')

with ProcessPoolExecutor(50) as p:
    for i in range(100):
        p.submit(func2(), key=f"线程{i}")
# 等待进程池中的任务全部执行完成，才继续执行
print('end')
```

## 13. 协程

```python
import asyncio
import time
```

```python
async def func4(key='abc', during=1):
    print(key)
    # time.sleep(3)
    await asyncio.sleep(3)
    print(key)
```

```python
async_func1 = func4(key='async 1', during=3)
async_func2 = func4(key='async 2', during=1)
async_func3 = func4(key='async 3', during=2)

start = time.time()
asyncio.run(asyncio.wait([async_func1, async_func2, async_func3]))
print(time.time() - start)
```
